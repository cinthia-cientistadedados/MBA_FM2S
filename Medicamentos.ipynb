flitrados

```python
df_filtrado = df.dropna()
df_filtrado
```

```python
# @title Nome do Medicamento

from matplotlib import pyplot as plt
import seaborn as sns
df_filtrado.groupby('Nome do Medicamento').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Valor

from matplotlib import pyplot as plt
import seaborn as sns
df_filtrado.groupby('Valor').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

Number of Medications by Manufacturer (Número de medicamentos por Fabricante)

```python
from matplotlib import pyplot as plt
import seaborn as sns
df_filtrado.groupby('Fabricante').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
