Integraremos o **Matplotib** e o **Seborn** para criar visualizações gráficas que nos ajudarão a entender melhor os padrões e tendêmcias nos daods. Utilizaremos os gráficos de barrras, histogramas de dispersão ou outros tipos de gráficos, conforme apropriado para representar diferentes aspectos dos dados de nomes do IBGE

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carregando dados da tabela funcionarios sep ;
df = pd.read_csv('funcionarios.csv', sep=';')
df
```

```python
#histograma usando Matplotlib
plt.figure(figsize=(8,6))
plt.hist(df['Idade'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribuição de idade')
plt.xlabel('Idade')
plt.ylabel('Frequencia')
plt.show()
```

```python
#Boxplot usando Seaborn
plt.figure(figsize=(8,6))
sns.boxplot(x='Idade', y='Salario', data=df)
plt.xlabel('Idade')
plt.ylabel('Salario')
plt.show()
```
