```python
import pandas as pd
df = pd.read_excel('corrigindoVendas.xlsx', sheet_name = 'Planilha1')
df_corrigido = pd.melt(df,id_vars=['Data'], var_name='Sigla',value_name='Venda')
df_corrigido
```

```python
df_corrigido['Venda'] = pd.to_numeric(df_corrigido['Venda'],errors='coerce')
df_corrigido
```

```python
df_corrigido['Ano'] = df_corrigido['Data'].dt.year
df_corrigido
```

```python
df_estados = pd.read_excel('frete.xlsx')
df_estados
```

```python
df_procv = pd.merge(df_corrigido, df_estados[['Sigla', 'Região', 'Frete']], on='Sigla', how='left')
df_procv
```

```python
df_procv = df_procv.assign(Total=df_procv['Venda'] + df_procv['Frete'])
df_procv
```

```python
df_procv.to_csv('baseVendasCorrigida.csv', index=False)
```
