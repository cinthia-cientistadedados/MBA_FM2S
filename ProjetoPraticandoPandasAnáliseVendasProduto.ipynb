1. Criação do DataFrame

```python

import pandas as pd

dados = {
    'Produto': ['A', 'B', 'C', 'D'],
    'Vendas_Janeiro': [120, 340, 560, 230],
    'Vendas_Fevereiro': [150, 300, 620, 210]
}

df = pd.DataFrame(dados)
print(df)
```

2. Cálculo da Média de Vendas

```python

df['Media_Vendas'] = df[['Vendas_Janeiro', 'Vendas_Fevereiro']].mean(axis=1)
print(df[['Produto', 'Media_Vendas']])
```

3. Visualização com Gráfico de Barras

```python

import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.bar(df['Produto'], df['Media_Vendas'], color='skyblue')
plt.title('Média de Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Média de Vendas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
